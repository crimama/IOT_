0. YOLOv5 와 RepVGG 모델을 이용해 Emotional Detection을 진행 함 
1. Real time Emotion Detction은 Webcam을 사용하기 때문에 로컬에서 진행 됨 
2. 로컬에서 프롬프트를 킴 
3. Git Clone 
```
git clone  https://github.com/George-Ogden/emotion.git 
```
4. 디렉토리 변경 
```
 cd ./emotion 
```
5. 가상 환경 설치 
```
conda env create -f env.yaml
``` 
6. 가상 환경 활성화 
```
conda activate emotion 
```
7. 필요한 라이브러리 설치
```
pip install -r requirements.txt
```
8. python main.py 입력하면 webcam이 켜지며 emotion detection이 실행 됨 
```
python main.py
```
